## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

# My Lane as an ML Task

**Task Type:** Classification

The goal is to predict whether a delivery will result in a wicket.

Each delivery (ball) is treated as one observation.

Possible outcomes:
- 1 = Wicket
- 0 = No Wicket

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

# Target or Proxy

Target Column: `is_wicket`

Since the dataset does not contain a ready-made target column, we create one using the `wicket_type` column.

If `wicket_type` is present, then:
- 1 = Wicket
- 0 = No Wicket

## 3. Success metric

*One metric you can defend. What number means 'good'?*

# Success Metric

Primary Metric:
- F1 Score

Other Metrics:
- Precision
- Recall
- Accuracy

F1 Score is appropriate because wicket deliveries are much less common than non-wicket deliveries.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

# Why ML Beats a Fixed Rule

A simple rule cannot accurately predict whether a wicket will fall because many factors influence the outcome of a delivery, including:

- Bowler
- Batter
- Over number
- Batting team
- Bowling team
- Match situation

Machine Learning can learn complex patterns from historical data and estimate the likelihood of a wicket more accurately than fixed rules.
